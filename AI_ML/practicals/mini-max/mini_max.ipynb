```python
import math

def getTree(leafs):
    # Find the height of the tree
    k = len(leafs)
    total_nodes = k
    p = k
    while(p>1):
      p = math.ceil(p/2)
      total_nodes = total_nodes + p

    print(k, total_nodes)
    adj_list = {}
    for i in range(0, total_nodes):
      #print(i, end = " ")
      node = {'value':None, 'children':[]}
      if(i >= total_nodes-k):
        node['value'] = leafs[i - total_nodes]
      elif((2*i + 2) < total_nodes):
        node['children'] = [2*i + 1, 2*i + 2]
      else:
        node['children'] = [2*i + 1]
      adj_list[i] = node
    return(adj_list)


print(getTree([4, 5, 6]))
```

```python
class Minimax:
    def __init__(self, graph):
        self.graph = graph

    def mima(self, currentNode, maxTurn):

        # Base case: leaf node
        if self.graph[currentNode]['children'] == []:
            return self.graph[currentNode]['value']
        elif maxTurn:  # Maximizing player's turn
            # evaluate each child and select max value
            maxEval = -float('inf')
            for child in self.graph[currentNode]['children']:
              a = self.mima(child, False)
              maxEval = max(maxEval, a)
            return maxEval
        else:     # Minimizing player's turn
            # evaluate each child and select min value
            minEval = float('inf')
            for child in self.graph[currentNode]['children']:
              a = self.mima(child, True)
              minEval = min(minEval, a)
            return minEval

    def start(self):
        return self.mima(list(self.graph.keys())[0], True)
```

```python
adj_list = {
    'A': {'value': None, 'children': ['B', 'C']},
    'B': {'value': None, 'children': ['D', 'E']},
    'C': {'value': None, 'children': ['F', 'G']},
    'D': {'value': 40, 'children': []},
    'E': {'value': 50, 'children': []},
    'F': {'value': 60, 'children': []},
    'G': {'value': 70, 'children': []}
}

obj = Minimax(adj_list)
print(obj.start())
```
