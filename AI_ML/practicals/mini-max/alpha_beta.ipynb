if alpha >= beta : break

max : alpha = max(alpha, maxEval)

min : beta = min(beta, minEval)

a minima parent node stops processing its child after it encounters a child whose value <= parents's left sibling's node's value

a maxima parent node stops processing its child after it encounters a child whose value >=  parents's left sibling's node's value

```python
class AlphaBeta:
    def __init__(self, graph):
        self.graph = graph

    def albe(self, currentNode, maxTurn, alpha, beta):

        # Base case: leaf node
        if self.graph[currentNode]['children'] == []:
            return self.graph[currentNode]['value']
        elif maxTurn:  # Maximizing player's turn
            # evaluate each child and select max value
            maxEval = -float('inf')
            for child in self.graph[currentNode]['children']:
              a = self.albe(child, False, alpha, beta)
              maxEval = max(maxEval, a)
              alpha = max(alpha, maxEval)
              if alpha >= beta:
                print("prunned at parent node = ", currentNode, " and childs after node = ", child)
                break
            return maxEval
        else:     # Minimizing player's turn
            # evaluate each child and select min value
            minEval = float('inf')
            for child in self.graph[currentNode]['children']:
              a = self.albe(child, True, alpha, beta)
              minEval = min(minEval, a)
              beta = min(beta, minEval)
              if alpha >= beta:
                print("prunned at parent node = ", currentNode, " and childs after node = ", child)
                break
            return minEval

    def start(self):
        alpha = -float('inf')
        beta = float('inf')
        return self.albe(list(self.graph.keys())[0], True, alpha, beta)
```

```python
adj_list = {
    'A': {'value': None, 'children': ['B', 'C']},
    'B': {'value': None, 'children': ['D', 'E']},
    'C': {'value': None, 'children': ['F', 'G']},
    'D': {'value': None, 'children': ['H', 'I']},
    'E': {'value': None, 'children': ['J', 'K']},
    'F': {'value': None, 'children': ['L', 'M']},
    'G': {'value': None, 'children': ['N', 'O']},
    'H': {'value': 3, 'children': []},
    'I': {'value': 5, 'children': []},
    'J': {'value': 6, 'children': []},
    'K': {'value': 9, 'children': []},
    'L': {'value': 1, 'children': []},
    'M': {'value': 2, 'children': []},
    'N': {'value': 0, 'children': []},
    'O': {'value': -1, 'children': []}
}

obj = AlphaBeta(adj_list)
print("Final value at root = ", obj.start())
```
